# 挑戰題二【我有一棵樹】：

```python
def tree(x):
    if (x%1==0):
        for i in range(1,x+1):
            print((x-i)*' '+(2*i-1)*'*')
    else:
        print('請輸入整數!')
tree(6)
```

# 挑戰題三：自己出！【巴斯卡三角形】

```python
def pascal(k):
    if k == 1:
        print(1)
    elif (k > 1)&((k%1)==0):
        層 = [[1],[1,1]]
        if k > 3:
            for i in range(3,k+1):
                層內=[1]
                for j in range(0,i-2):
                    層內.append(層[i-2][j]+層[i-2][j+1])
                層內.append(1)
                層.append(層內)
    else:
        print('請輸入正整數!')
    last_string_len = len(" ".join(str(層[k-1][l]) for l in range(0,len(層[k-1]))))
    print(int((last_string_len-1)/2)*' '+str(層[0][0]))
    for i in range(1,k-1):
        tstr=" ".join(str(層[i][l]) for l in range(0,len(層[i])))
        print ((int((last_string_len-len(tstr))/2))*" "+tstr)
pascal(20)
```
