## 挑戰題一：買相機

```python
p1 = input("攝影設備: ")
```

```python
p2 = input("攝影設備: ")
```

```python
p3 = input("攝影設備: ")
```

```python
p4 = input("攝影設備: ")
```

```python
cp = [0, float(p1), float(p1)+float(p2), float(p1)+float(p2)+float(p3), float(p1)+float(p2)+float(p3)+float(p4)]
```

```python
cp
```

```python
plot(cp)
plt.xlabel('time')
plt.ylabel('prices')
plt.xticks([1, 2, 3, 4], ['$Next Month$', '$2M-later$', '$3M-later$', '$4M-later$'])
```

```python
labels = 'Next Month', '2M later', '3M later', '4M later'
sizes = [100, 150, 230, 80]
colors = ['lightgreen', 'gold', 'lightskyblue', 'lightpink']
explode = (0, 0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Cumulative Consumption of Cameras")
```
