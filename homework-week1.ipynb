挑戰題一：買相機

財團法人大兔兔文教基金會要添購攝影設備。
他們預計在未來四個月內分別添購四項價格為 100k,150k,230k,80k的設備，
請以折線圖畫出累積消費額，並以圓餅圖畫出各器材所佔比例。

```python
a = 100
b = a + 150
c = b + 230
d = c + 80
plot([1,2,3,4],[a,b,c,d])
```

財團法人大兔兔文教基金會未來四個月添購攝影設備所花累積費用。 （橫軸：未來X月/ 縱軸：累積花費（千））

```python
labels = 'month1', 'month2', 'month3', 'month4'
fracs = [100, 150, 230, 80]
explode=(0, 0, 0, 0)

pie(fracs, explode=explode, labels=labels,
                autopct='%1.1f%%', shadow=True, startangle=90)
```
