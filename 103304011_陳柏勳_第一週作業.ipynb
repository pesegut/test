## 第二題：台北市最高最低溫分佈圖
* **題目說明：請以plot函式畫出台北市近年來的高低溫分佈圖。**

```python
最高溫=[19.1,19.6,22.1,25.7,29.2,32.0,34.3,33.8,31.1,27.5,24.2,20.7]
最低溫=[13.9,14.2,15.8,19.0,22.3,24.6,26.3,26.1,24.8,22.3,19.3,15.6]
x = range(1,13)
plot(x,最高溫,'r-.',x,最低溫,'b-.')
plot(x,最高溫,'ro',x,最低溫,'bo')
legend(['Highest Temperature','Lowerest Temperature'])
title('''Average highest and lowerest temperature
between 1981 and 2010 in Taipei''')
xticks(range(1,13))
xlabel("month")
ylabel("temperature(℃)")
show()
```

## 挑戰題一：買相機
* 題目說明：財團法人大兔兔文教基金會要添購攝影設備。
    他們預計在未來四個月內分別添購四項價格為     100k,150k,230k,80k的設備，
    請以折線圖畫出累積消費額，並以圓餅圖畫出各器材所佔比例。 

```python
消費額=[100,150,230,80]
累積消費額=[消費額[0]]
for i in range(1,4):
    累積消費額.append(累積消費額[i-1] + 消費額[i])
figure(1)
plot(range(1,5),累積消費額,'r-.')
plot(range(1,5),累積消費額,'ro')
xticks(range(1,5))
ylim(80,580)
xlabel("month")
ylabel("cumulative cost")
figure(2)
x = []
for i in range(1,5):
    x.append("Equipment "+str(i))
pie(消費額,labels=x)
show()
```

* **嘗試畫出常態分配圖**

```python
def nf(x,mu,sigma):
    pdf = exp(-((x - mu)**2) / (2* sigma**2)) / (sigma * np.sqrt(2*np.pi))
    return pdf
data = randn(100)
```

```python
x = linspace(-5,5,100)
y = nf(x,0,1)
plot(x,y, color='r')
hist(data,bins=10,rwidth=0.5,normed=True)
title('Normal Distribution')
xlabel('x')
ylabel('Probability')
show()
```
